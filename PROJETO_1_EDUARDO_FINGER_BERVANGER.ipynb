```python
print('Bem vindo ao primeiro projeto do curso: A calculadora')
print('Escolha uma das fórmulas a seguir: Adição, Subtração, Multiplicação, Divisão, Exponenciação')
fórmulas_matemáticas=input()
if fórmulas_matemáticas=='Adição':
  adi1=(int and float(input('Escreva qual valor deseja fazer Adição: ')))
  adi2=(int and float(input('Escreva qual valor deseja fazer Adição: ')))
  print('Resultado:', adi1+adi2)
if fórmulas_matemáticas=='Subtração':
  sub1=(int and float(input('Escreva qual valor deseja fazer Subtração: ')))
  sub2=(int and float(input('Escreva qual valor deseja fazer Subtração: ')))
  print('Resultado:', sub1-sub2)
if fórmulas_matemáticas=='Multiplicação':
  mult1=(int and float(input('Escreva qual valor deseja fazer Multiplicação: ')))
  mult2=(int and float(input('Escreva qual valor deseja fazer Multiplicação: ')))
  print('Resultado:', mult1*mult2)
if fórmulas_matemáticas=='Divisão':
  div1=(int and float(input('Escreva qual valor deseja fazer Divisão: ')))
  div2=(int and float(input('Escreva qual valor deseja fazer Divisão: ')))
  print('Resultado:', div1//div2)
if fórmulas_matemáticas=='Exponenciação':
  exp1=(int and float(input('Escreva qual valor deseja fazer Exponenciação: ')))
  exp2=(int and float(input('Escreva qual valor deseja fazer Exponenciação: ')))
  print('Resultado:', exp1**exp2)
```
